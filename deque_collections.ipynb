```python
from collections import deque

dq = deque(list('asdfghj'),maxlen=19)
dq.append('hi')
dq
```

```python
#appends as written
dq.appendleft('ss')
dq
```

```python
# makes copy
copy_dq = dq.copy()
copy_dq
```

```python
#counts elem in dequie
dq.count('s')
```

```python
# extends.   --- EXTENDLEFT the same but does it on left side
dq.extend(list('hello'))
dq
```

```python
# returns position of value 'hi' works with start and stop position to find
start_pos = 0  # START_POSITION
stop_pos = 10  # STOP_POSITION
dq.index('hi',start_pos,stop_pos)
```

# USAGE PATTERNS FOR DEQUE

```python
from collections import deque

def tail(filename, n=10):
    "Вернуть последние 'n' строк файла"
    with open(filename) as f:
        return deque(f, n)
```

```python
import itertools
mm = deque([1,2,3,4,5,6,7,8,9])
def moving_average(iterable, n=3):
    "moving_average([40, 30, 50, 46, 39, 44]) --> 40.0 42.0 45.0 43.0"
    it = iter(iterable)
    d = deque(itertools.islice(it, n-1))
    d.appendleft(0)
    s = sum(d)
    for elem in it:
        s += elem - d.popleft()
        d.append(elem)
        yield s / n
        
    
    
```

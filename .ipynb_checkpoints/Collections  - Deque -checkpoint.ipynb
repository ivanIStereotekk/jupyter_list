### Collections Deque and its methods

#### Doubly Ended Queue - short deque
#docs are here - https://docs.python.org/3/library/collections.html#collections.deque

```python
from collections import deque

my_deq = deque(['hello','my','lovely','deque',[1,2,3]])
### deq
my_deq.append('appended')
print('deq is:',my_deq)
# appendleft
my_deq.appendleft('left_appended')
print('here we did appendleft:',my_deq)
# popped
popped1 = my_deq.pop()
print('pop method:',popped1)
# popleft
popped2 = my_deq.popleft()
print('popleft method',popped2)
print('the left deque is:',my_deq)
```

```python
deq2 = deque([88,88,1,2,3,4,5,88,6,6,6])

# deque INSERT  - first arg is: slice.   Second arg is: item to insert
position = 4
insertion = [777,888,999]
deq2.insert(position,insertion)
deq2
```

```python
### deque INDEX

deq3 = deque([1,2,3,4,5,6,7,'a','b','c',9,8,7,6,5])

#INDEX - works with 3 args: elem >>> beggining >>> end

elem_to_find = 'b'
start_search = 0
end_search = 14

#Returns index of searched element
deq3.index(elem_to_find, start_search, end_search)




# COUNT - counts elements in arg
deq3.count(7) 


# REMOVE - removes first occurances

deq3.remove(7)

deq3


# SIZEING DEQUE

len(deq3)
```

```python
import string
import random
# EXTEND >>> REVERSE >>> ROTATE

deq4 = deque([1,2,3,4])

# EXTEND takes itterable as arg for extend deque
it_ext = [i for i in string.punctuation if ord(i) > 100]
print("This is extention object:", it_ext)


deq4.extend(it_ext)
print('extended deq4 is:',deq4)

# Rotate - Rotates the deque , param is arg(int) 
#If negative arg (-99), rotation occurs to the left

nw = deque(['A',7,'>>>'])

print('rotated minus is :', nw.rotate(-1)) ### B U G working


# Revers - reversing deque
rr = nw.reverse()
    ### BUG WORKING
print('reversed is:',rr)
```

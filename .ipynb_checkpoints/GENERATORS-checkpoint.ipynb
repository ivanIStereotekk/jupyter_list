```python
def ss():
    while True:
        item = yield
        print('got some:',item)
        
        
s = ss()
next(s)
s.send('hello')
```

```python
ls1 = [1,2,3,4]
ls2 = [5,6,7,8]

def chain(x,y):
    yield from x
    yield from y
    

for one in chain(ls1,ls2):
    print(one)
```

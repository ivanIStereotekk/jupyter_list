```python
def base_func():
    aa = 8 # It's important to have correct names for nonlocals 
    def inner():
        nonlocal aa # When we define nonlocals we do these like one step upper scope
        xx = (aa + 100)
        print(xx)
    inner()
    global glob # When we do globals into any place we do these like global scope
    glob = "Created Global !"

base_func()
print(glob)
```

```python
# PATTERN LIKE COUNTER - behaves like a class

def counter(count=0):
    def step():
        nonlocal count
        count +=1
        return count
    return step

counter_1 = counter()
counter_2 = counter()
counter_3 = counter(500)

print(counter_1())
print(counter_1())
print(counter_1())
print(counter_2())
print(counter_2())
print(counter_3())
```

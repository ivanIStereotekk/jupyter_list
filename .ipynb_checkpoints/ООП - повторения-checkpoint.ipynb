```python
# method CALL определенный в классе делает его калбэчным

class Car(object):
    def __init__(self,a,b):
        self.a = a
        self.b = b
    def __call__(self):
        print("I can work as an Callback !")
        
car = Car(1,2)
car()
```

```python
# D U N D E R    M E T H O D S
# __eq__() – для равенства ==
# __ne__() – для неравенства !=
# __lt__() – для оператора меньше <
# __le__() – для оператора меньше или равно <=
# __gt__() – для оператора больше >
# __ge__() – для оператора больше или равно >=
```

```python
class Basic:
    def __init__(self,one, two):
        self.one = one
        self.two = two
        self.data = {}
    def __len__(self):
        return len(self.data)
    
    def __eq__(self,obj,two): # EQUALS
        if obj is None or not isinstance(obj,Basic):
            return False
        return obj.one == self.one
    
    def __ne__(self,one: int,two: int): # NOT EQUALS
        return self.one != self.two
    
    def __lt__(self,one: int,two: int): # LESS THAN
        return self.one < self.two
    
    def __le__(self,one: int,two: int): # LESS OR EQUALS
        return self.one == self.two
    
    def __gt__(self,one: int,two: int): # GREATER THAN
        return self.one >= self.two
    
bb = Basic(4,8)
bb.__eq__(1,1)
```
